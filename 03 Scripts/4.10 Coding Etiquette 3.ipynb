# Aggregate the max, mean, and min variables on a customer-profile level for usage frequency and expenditure
    # Customer profiles include: age_group, income_distribution, family_profile, department

```python
# creating age group agregate for order and price
age_group_agg = Normal_activity_customer.groupby('age_group').agg({'order_number': ['mean', 'min', 'max'], 'prices': ['mean', 'min', 'max']})
```

```python
age_group_agg
```

```python
# creating income_distribution agregate for order and price
income_distribution_agg = Normal_activity_customer.groupby('income_distribution').agg({'order_number': ['mean', 'min', 'max'], 'prices': ['mean', 'min', 'max']})
```

```python
income_distribution_agg.head()
```

```python
# creating family_profile agregate for order and price
family_profile_agg = Normal_activity_customer.groupby('family_profile').agg({'order_number': ['mean', 'min', 'max'], 'prices': ['mean', 'min', 'max']})
```

```python
family_profile_agg.head(8)
```

```python
# creating department agregate for order and price
department_agg = Normal_activity_customer.groupby('department').agg({'order_number': ['mean', 'min', 'max'], 'prices': ['mean', 'min', 'max']})
```
